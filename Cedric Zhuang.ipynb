#### At last, 多图
对各大洲的GDP,预期寿命及人口作图不同的子图

```python
import matplotlib.pyplot as plt
columns = ['lifeExp', 'pop', 'gdpPercap']
```

```python
fig, axes = plt.subplots(nrows=len(columns), sharex=True)

for i, col in enumerate(columns):
    grouped = nations[['continent', col, 'year']].groupby(['continent', 'year'])
    cont = grouped.sum() if col == 'pop' else grouped.mean()
    cont = cont.unstack().T.reset_index()
    cont.plot(ax=axes[i], figsize=[14, 16], x='year',style='.-', title='{} by contenant'.format(col), kind='line')
```
